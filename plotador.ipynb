```python
import matplotlib.pyplot as plt

# --- DADOS DO GRÁFICO ---
# Altere os valores abaixo para atualizar seu gráfico
# Os valores do Eixo X (primeira lista)
x_values = [0.029, 0.375, 0.508, 0.549, 0.577, 0.612, 0.619]

# Os valores do Eixo Y (segunda lista, correspondendo aos valores de X)
y_values = [0, 0.101, 0.205, 0.495, 0.907, 1.993, 2.428]


# --- CONFIGURAÇÕES DO GRÁFICO ---
# Você pode ajustar esses parâmetros para mudar a aparência do gráfico

# Define o tamanho da figura (largura, altura em polegadas)
plt.figure(figsize=(9, 6))

# Plota os dados:
# 'marker='o'' adiciona marcadores de círculo em cada ponto
# 'linestyle='-'' conecta os pontos com uma linha contínua
# 'color='purple'' define a cor da linha e dos marcadores
plt.plot(x_values, y_values, marker='o', linestyle='-', color='purple')

# Define o título do gráfico
plt.title('Meu Gráfico Personalizado de X vs. Y')

# Define o rótulo para o eixo X
plt.xlabel('Eixo X (Unidade)') # Altere 'Unidade' conforme necessário

# Define o rótulo para o eixo Y
plt.ylabel('Eixo Y (Unidade)') # Altere 'Unidade' conforme necessário

# Adiciona uma grade ao fundo do gráfico para facilitar a leitura
plt.grid(True)

# --- EXIBIÇÃO DO GRÁFICO ---
# Mostra a janela do gráfico
plt.show()
```
